```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Conhecendo os dados

```python
dados = pd.read_csv('imigrantes_canada.csv')
```

```python
dados.head()
```

```python
dados.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = dados.loc['Brasil', anos]
```

```python
brasil
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes' : brasil.values.tolist()}
```

```python
brasil_dict
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

# Adicionando títulos e rótulos
# Criando o primeiro gráfico com Matplotlib

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canada')
plt.xlabel('Ano')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.ylabel('Numero de imigrantes')
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])

plt.show()
```

# Criando uma figura
# Adicionando títulos e rótulos em figuras

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para canada \n1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Número de imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show
```

# Criando subplots em uma direção

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canada\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Imigração')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot imigração do Brasil para o Canada\n 1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Imigração')
axs[1].grid()

plt.show()
```

# Criando subplots em duas direções #1 e #2


```python
dados_brasil.describe()
```

```python
fig, axs = plt.subplots(2,2, figsize=(11,7))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigração dos quatros maiores paises da Ámerica do sul para Canadá\n 1980 a 2013 ')

axs[0,0].plot(dados.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(dados.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(dados.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(dados.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    
for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Imigrantes')    

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

# Alterando as cores #1

```python
dados.head()
```

```python
america_sul = dados.query('Região == "América do Sul"')
```

```python
america_sul
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid',
         'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']


fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul.index, america_sul['Total'], color=cores)
ax.set_title('Imigração da América do Sul par o Canada\n1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

# Alterando as cores #2
# Adicionando anotações

```python
america_sul_ordenado = america_sul.sort_values('Total', ascending=True)
```

```python
cores = []

for pais in america_sul_ordenado.index:
    if pais == 'Brasil':
      cores.append('green')
    else:
       cores.append('silver')  
     

fig, ax = plt.subplots(figsize=(12, 5))
ax.barh(america_sul_ordenado.index, america_sul_ordenado['Total'], color=cores)
ax.set_title('America do Sul: Brasil foi quarto pais com mais imigrantes \n para o Canada no período de 1980 a 2013', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_ordenado['Total']):
 ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

fig.savefig('Imigracao_America_Sul.png',transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

# Personalizando visualizações com Seaborn

```python
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h')

ax.set(title='Países com maior imigração para o Canada\n 1980 a 2013,',
       xlabel='Número de imigrantes',
       ylabel='')
plt.show
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=top_10, y=top_10.index,
                 x='Total', orient='h', palette='tab10_r')

ax.set_title('Paéses com maior imagração para o Canada \n1980 a 2010',
             loc='left', fontsize=16)
ax.set_xlabel('Némero de imigrantes', fontsize=14)
ax.set_ylabel('')

plt.show
```

# Alterando as cores com paletas

```python
def gerar_grafico_palleta(pallete):
    fig, ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=pallete)
    ax.set_title('Países com maior imigração para o Canada \n1980 a 2010', loc='left', fontsize=16)
    ax.set_xlabel('Número de imigrantes', fontsize=14)
    ax.set_xlabel('')

    plt.show()
```

```python
gerar_grafico_palleta('Blues')
```

```python
gerar_grafico_palleta('Blues_r')
```

```python
gerar_grafico_palleta('rocket')
```

```python
gerar_grafico_palleta("tab10")
```

# Explorando temas

```python
sns.set_style('dark')
gerar_grafico_palleta('tab10')
```

```python
sns.set_theme(style='whitegrid')
gerar_grafico_palleta('tab10')
```

```python
sns.set_theme(style='white')
gerar_grafico_palleta('tab10')
```

```python
sns.set_theme(style='ticks')
gerar_grafico_palleta('tab10')
```

## Alterando as cores com Plotly

```python
import plotly.express as px
```

```python
fig = px.line(dados_brasil, x='ano', y='imigrantes',
              title='Imigração do Brasil para Canada no período de 1980 a 2013')
fig.update_traces(line_color='green', line_width=4)
fig.update_layout(width=1000, height=500, 
                  xaxis={'tickangle':-45}, 
                  font_family='Arial',
                  font_size=14,
                  font_color='grey',
                  title_font_color='black',
                  title_font_size=22,
                  xaxis_title='Ano',
                  yaxis_title='Número de imigrantes')


fig.show()
```
